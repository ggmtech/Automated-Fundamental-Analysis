```python
start = date.today() - relativedelta(years=5)
end = date.today()

def bollinger_bands(df, window=20):
    df[f'SMA(window={window})'] = df['Close'].rolling(window=window).mean()
    df[f'STD(window={window})'] = df['Close'].rolling(window=window).std()

    df[f'Upper Band{window}'] = df[f'SMA(window={window})'] + (df[f'STD(window={window})'] * 2)
    df[f'Lower Band{window}'] = df[f'SMA(window={window})'] - (df[f'STD(window={window})'] * 2)
    return df

def create_rsi(df, window=14):
    delta = df['Close'].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(window=window).mean()
    avg_loss = loss.rolling(window=window).mean()
    rs = avg_gain / avg_loss
    rsi_value = 100 - (100 / (1 + rs))
    df['RSI'] = rsi_value
    return df

def money_flow_index(df, window=14):
    typical_price = (df['High'] + df['Low'] + df['Close']) / 3
    raw_money_flow = typical_price * df['Volume']
    positive_flow = raw_money_flow.where(typical_price > typical_price.shift(1), 0)
    negative_flow = -raw_money_flow.where(typical_price < typical_price.shift(1), 0)

# calculate money flow ratio
    window = 14
    positive_mf_sum = positive_flow.rolling(window=window).sum()
    negative_mf_sum = negative_flow.rolling(window=window).sum()
    money_flow_ratio = positive_mf_sum / negative_mf_sum

    mfi = 100 - (100 / (1 + money_flow_ratio))
    df['mfi'] = mfi
    return df

```

```python
# custom function with trading algo
def algorithm(ticker, start = start, end = end, interval= '1d'):
    df = yf.download(ticker, start, end = end, interval= interval)
    df.columns = [col[0] for col in df.columns]

    df = bollinger_bands(df=df, window=12)
    df = bollinger_bands(df=df, window=20)
    df = bollinger_bands(df=df, window=50)

    df = create_rsi(df=df, window=14)
    df = money_flow_index(df=df, window=14)


    # create logic for buy and sell
    ## initialise values
    low= df['Low']
    high = df['High']
    sma12 = df['SMA(window=12)']
    sma20 = df['SMA(window=20)']
    sma50 = df['SMA(window=50)']
    bollingerupper20 = df['Upper Band20']
    bollingerlower20 = df['Lower Band20']
    bollingerupper50 = df['Upper Band50']
    bollingerlower50 = df['Lower Band50']
    rsi = df['RSI']
    mfi = df['mfi']

    buy_signal = []
    sell_signal = []
    action = []


    for i in range(len(df)):
        low_value = low[i]
        high_value = high[i]
        sma12_value = sma12[i]
        sma20_value = sma20[i]
        sma50_value = sma50[i]
        bollingerlower_val20 = bollingerlower20[i]
        bollingerupper_val20 = bollingerupper20[i]
        boliingerlower_val50 = bollingerlower50[i]
        bollingerupper_val50 = bollingerupper50[i]
        rsi_value = rsi[i]
        mfi_value = mfi[i]

        # buy logic
        buy_condition1 = (low_value < bollingerlower_val20) and (rsi_value < 30 and mfi_value < 25 )
        sell_condition1 = (high_value > bollingerupper_val20) and (rsi_value > 70 and mfi_value > 75)

        # scatter plot values for buy and sell signal
        if buy_condition1:
            buy_signal.append(low_value*0.95)
            sell_signal.append(np.nan)
            action.append('buy')

        elif sell_condition1:
            buy_signal.append(high_value*1.05)
            sell_signal.append(high_value)
            action.append('sell')
        else:
            buy_signal.append(np.nan)
            sell_signal.append(np.nan)
            action.append('hold')

    df['buy_signal'] = buy_signal
    df['sell_signal'] = sell_signal
    df['action'] = action


    #Back testing logic
    initial_cash = 100000
    cash = initial_cash
    position = 0
    portfolio_value = []

    # buy and hold
    share_cost = df['Open'].iloc[0]
    num_shares = initial_cash / share_cost
    df['Buy_Hold_Value'] = df['Close']*num_shares


    # iterate over each row
    for i in range(0, len(df)):
       action = df['action'].iloc[i]
       price = df['Close'].iloc[i]

       # if action is buy
       if action == 'buy' and cash > 0:
           position = cash / price
           cash = 0
       # if action is sell
       elif action == 'sell' and position > 0:
           cash = position * price
           position = 0

       elif action == 'Hold':
            pass

       portfolio_value.append(cash + position*price) # portfolio value at end of the day

    df['portfolio_value'] = portfolio_value
    return df

```
